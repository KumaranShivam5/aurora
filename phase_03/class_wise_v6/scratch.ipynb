```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("processed_data/train_norm_median_impute")
df
```

```python
f = lambda x: 1 if x=='NS' else 0
l = list(map(f,df['class']))
```

```python
plt.scatter(x=df['flux_aper_hilim_s'] , y=df['flux_aper_hilim_u'] , c=l)
```

```python
y =  df['class']
x = df.drop(['index', 'class' ,'src_id' ,'src_n' ,'significance'] , axis=1)
display(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
X_train.shape
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
demo = pd.DataFrame()
demo['columns'] = X_train.columns
demo['weights'] = model.coef_[0]
demo = demo.sort_values(by='weights')
demo
```
